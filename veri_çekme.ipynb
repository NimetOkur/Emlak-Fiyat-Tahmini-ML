```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup as bs  
```

```python
link_list = []
prices_list = []
titles = []
values = []

for i in range(1,25): 
    url = "https://www.emlakjet.com/satilik-konut/manisa/" + str(i)  # Url'in for döngüsü ile güncellenmesi
    r = requests.get(url)  #url'e istek atılması
    source = bs(r.content, "html.parser")  # İstek yapılan sayfanın içeriğinin çekilmesi
    
    
    urls = source.find_all("div", attrs={"class": "_3qUI9q"})  # Tüm veriler
    for url in urls:
        house_url = "https://www.emlakjet.com/" + url.a.get("href")  # Verilerin linklerinin bulunması
        link_list.append(house_url)  # Linklerin kaydedilmesi
        print(house_url)
        
        r_house = requests.get(house_url)  # Bulunan linke tekrar istek atılması
        source_house = bs(r_house.content, "html.parser")  # Veri içeriğinin çekilmesi

        properties = source_house.find_all("div", attrs={"class": "_35T4WV"})  # tüm özelliklerin bulunması
        for prop in properties:
            title = prop.contents[0].text.strip()
            value = prop.contents[1].text.strip()
            titles.append(title)
            values.append(value)
            
            
             
    prices = source.find_all("p", attrs={"class": "_2C5UCT"})  # Fiyatların bulunması
    for price in prices:
        prices_list.append(price.text)  # Fiyatların listeye eklenmesi
        
```

```python
print(str(len(link_list)) + " Adet link bulundu.")
print(str(len(prices_list)) + " Adet fiyat bulundu.")
print(str(len(titles)) + " Adet özellik başlığı bulundu.")
print(str(len(values)) + " Adet özellik verisi bulundu.")
```

```python
data = {
    
    "url": link_list,
    "price": prices_list
}
df = pd.DataFrame(data)
df.head()
```

```python
#bu aşamayı başka şekilde yap
data = {
    
    "url": link_list,
    "price": prices_list
}

for title, value in zip(titles, values):
    data[title] = value

df = pd.DataFrame(data)
```

```python
houses = len(link_list)
# Her bir URL için
for i in range(0, houses):
    url = df['url'].loc[i]
    request = requests.get(url)
    source = bs(request.content, "html.parser")
    # Her bir özellik için
    properties = source.find_all("div", attrs={"class": "_35T4WV"})  # tüm özelliklerin bulunması
    for prop in properties:
        title = prop.contents[0].text.strip() 
        value = prop.contents[1].text.strip()
        titles.append(title)
        values.append(value)
        df[title].loc[i] = value
    
      
```
